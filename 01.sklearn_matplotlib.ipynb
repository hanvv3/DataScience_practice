```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
sampleNum = 5
img = digits.images[sampleNum]

plt.figure(figsize=(8, 2))
plt.imshow(img, interpolation='nearest', cmap=plt.cm.bone_r)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("image {}".format(sampleNum+1))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
samples = [0, 10, 20, 30, 1, 11, 21, 31]
d = []
lstlen = len(samples)

for i in range(lstlen):
    d.append(digits.images[samples[i]])

#    (using numpy)
# D = np.array(d)
# print(D.shape)

plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("image {}".format(i + 1))

plt.suptitle("digit 0 & 1 image")
plt.tight_layout()
plt.show()
```

```python
#   with 'np'
D = np.array(d)
print(D.shape)
```

```python
v = []
for i in range(8):
    v.append(d[i].reshape(64,1))
    
V = np.array(v)
print(V.shape)
```

```python
plt.figure(figsize=(8,3))

for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(v[i], aspect=0.2, interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("vector {}".format(i + 1))

plt.suptitle("vector image")
plt.tight_layout(w_pad=7)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import misc

#   
from matplotlib import rc
%matplotlib inline

plt.rcParams['axes.unicode_minus'] = False

rc('font', family='AppleGothic', size=12)
# 

#    
img_rgb = misc.face()
img_rgb.shape

# 2x2 1 
plt.subplot(221)
plt.imshow(img_rgb, cmap=plt.cm.gray)
plt.axis("off")
plt.title("RGB  ")

plt.subplot(222)
plt.imshow(img_rgb[:, :, 0], cmap=plt.cm.gray)
plt.axis("off")
plt.title("Red ")

plt.subplot(223)
plt.imshow(img_rgb[:, :, 1], cmap=plt.cm.gray)
plt.axis("off")
plt.title("Green ")

plt.subplot(224)
plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.gray)
plt.axis("off")
plt.title("Blue ")

plt.show()
```
