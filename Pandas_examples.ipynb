### DataFrame 생성하기

```python
# 방법1: list로 만들기
company1 = [['삼성',2000,'스마트폰'],
           ['현대',1000,'자동차'],
           ['네이버',500,'포털']]
```

```python
# 자동으로 테이블을 생성
df1 = pd.DataFrame(company1)
df1
```

```python
# 컬럼이름을 인덱스 순서대로 넣어준다.
df1.columns = ['기업명','매출액','대표제품']
df1
```

```python
# 방법2: dict로 만들기
company2 = {'기업명':['삼성','현대','네이버'],
            '매출액':[2000,1000,500],
            '대표제품':['스마트폰','자동차','포털']
           }
```

```python
df2 = pd.DataFrame(company2)
df2
```

## loc

- Pandas에서 loc조건을 줄땐, numpy와 달리 슬라이싱 인덱스의 stop은 "이하"이기 때문에 아래의 10번째 인덱스 까지 포함한다.

```python
# 모든 행의 2014년 데이터
df.loc[5:10,'2014년']
```

```python
needCol = ['기관명','2014년','2015년']

df.loc[6:9,needCol]
```

```python
# 이렇게도 사용가능

df.loc[6:9,'2014년':'2016년']
```

## iloc 색인

- i: index값으로 접근, 이름대신 이름으로 접근한다.
- 여기선 loc과 다르게 stop값을 미만으로 정의하기 때문에, stop 인덱스의 값을 포함하지 않는다.

```python
df.head()
```

```python
# column명 대신 번호를 사용한다.

df.iloc[:,[0,3]]
```

```python
# 1번째 행 - 3번째 행 미만
# 0번쨰 열 - 3번째 열 미만
df.iloc[1:3,0:3]
```

```python
# Boolean indexing으로 filter(검색조건)을 적용하여 소계의 값이 1000미만인 값만 가져오기.
df[df['소계']<1000]
```

```python
df[df['소계']<1000][['기관명','2013년도 이전','2014년']]
```

## 데이터프레임에 행 추가

```python
# dict 타입
copy_df.append({'기관명':'광진구2','소계':420,'2013년도 이전':100, '2014년':100, '2015년':100,'2016년':120}, ignore_index=True)

```

```python
# column 추가:
copy_df['2021년'] = 150

copy_df.head()
# 단점: 전부 동일한 값이 추가된다.
```

- 단점: 전부 동일한 값이 추가된다.

```python
# 데이터 사이에 행 추가하기 (0과 1사이)
new_data = {'기관명':'광진구2',
            '소계':420,
            '2013년도 이전':100,
            '2014년':100, 
            '2015년':100,
            '2016년':120
           }

# 원하는 위치
idx = 1
temp1 = copy_df[copy_df.index<idx]
temp2 = copy_df[copy_df.index>=idx]


copy_df = temp1.append(new_data, ignore_index=True).append(temp2)

copy_df.head()
```
